```python
tr18=pd.read_csv('tracking2018.csv.zip')
```

```python
plays=pd.read_csv('plays.csv.zip')
```

```python
#gameId, playId
merged=pd.merge(tr18,plays,on=['gameId','playId'])
```

```python
def printCols(data):
    n=0
    for i in data.columns:
        print(i,n)
        n=n+1
```
